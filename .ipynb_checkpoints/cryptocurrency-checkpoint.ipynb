```python
df_platforms =pd.DataFrame()
for index,row in coin_assetplatform_df.iterrows(): 
    df={}
    df['id']= row['id']
    df['symbol']=row['symbol']
    df['name']=row['name']
    for item in row['platforms']:
        if item is not None:
            df['platforms']=item
            df = pd.DataFrame(df, index=[0])
            df_platforms = df_platforms.append(df) 
```
